##문제2

```python
numbers = [1,2,6,8,4,3,2,1,9,5,4,9,7,2,1,3,5,4,8,9,7,2,3]
counter = {}
```

```python
# counter에 각 숫자를 몇번이나 확인하였는지
# 값이 있다면 누적해서 아니라면 key를 만들고 1로 확인
for number in numbers:
    if (counter.get(number)):
       counter[number] += 1
    else:
        counter[number] =1
print(counter)
```

```python
for number in numbers:
    if number not in counter :
        counter[number] = 1
    else :
        counter[number] = counter[number]+1

print(counter)
```

```python
character = {
    "name": "기사",
    "level" : 12,
    "items" : {
        "sword":"불꽃의 검",
        "armor": "풀플레이트"
    },
    "skill": ["베기","세게 베기","아주 세게 베기"]
}
character
```

```python
i=0
for key,value in character.items() :
    print(f"{key} : {value}")
    # value가 dict인 경우
    # print(type(val))
    if type(value) is dict :
        for sub_key, sub_val in value.items():
            print(f"{sub_key} : {sub_val}")
    if type(value) is list :
        for l_value in value:
            i +=1
            print(f"skill{i} : {l_value}")
```
