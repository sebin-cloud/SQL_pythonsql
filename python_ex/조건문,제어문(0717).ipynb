```python
#문제1
# if 돈이 있거나 or 카드가 있거나
# 둘다 없을 경우 > 걸어가기
pocket = "money, paper, mobilepaper"
card = True
if "money" in pocket :
    print("택시를 타세요")
elif "money" not in pocket and card == True :
    print("택시를 타세요")
elif "money" not in pocket and card == False :
    print("걸어가세요")
```
