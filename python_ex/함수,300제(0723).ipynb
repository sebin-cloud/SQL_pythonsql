# 제어문 문제3

```python
#알고리즘 부터 먼저 짜기
# 1~100 사이 소수 구하기 -> 1과 자기자신만을 약수로 가지는 수 (약수가 없음)
# 2~n-1의 숫자로 자신을 나눴을 때, 나눠지는 수가 없다면 소수
decimal = []
decimal.append(2)
for n in range(2,101):
        #분모 : denominator
    deno = n-1
    while (deno>2):
        #n=3이라면 판단방법
        if n % deno == 0 :
            break
        if deno == 2:
            decimal.append(n)
        deno = deno -1
print(decimal)

```
