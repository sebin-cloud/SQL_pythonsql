```python
cnt = 1
while cnt <= 5 : # cnt가 5보다 같거나 작을 동안 반복
    print(f"{cnt} : 윤세빈")
    cnt = cnt +1
```

```python
import time
number = 0

target_tick = time.time()+5
while time.time() <  target_tick :
    number += 1

print("5초 동안 {}번 반복했습니".format(number))
```

```python
i = 0
while True :
    print("{}번째 반복문입니다.".format(i))
    i = i + 1
    input_text = input("> 종료하시겠습니까?(y/n):")
    if input_text in ["y","Y"]:
        print("반복을 종료합니다.")
        break
```

```python
#누적 곱, 처음으로 10000을 넘어가는 i
limit = 10000
i = 1
#누적곱 변수 초기화
sum_val = 1

while sum_val <= limit : #조건이 참일 동안 반복
    sum_val = sum_val * i
    i = i + 1
print(f"{limit}을 넘는 i는 {i}이고 합값은 {sum_val}입니다.")
```
