```python
# for in data (컨테이너형,반복할 수 있는 것을 넣어야함)
data = [11,22,33,'a','b','c']
```

```python
# 변수 a 와 위의 'a'는 다른 것 / a는 각 11,22,33,'a','b','c'가 됨
index = 0
for a in data :
    print(f"인덱스 {index},  값 {a}")
    index = index+1
    print("반복문 안에 있음")
    print("-"*20)
print("반복문 밖이야")
```

##문제3

```python
input에 해당 조건을 넣기
1의 자리일 경우 > 바로 넣으면 됨

2의 자리일 경우 > 10으로 나눠서 넣고 - 목 *10
3의 자리일 경우 > 100으로 나누고 10으로 나누고 빼기

```

```python
a = input("1에서 1000사이에 있는 숫자를 입력해주세요")
a = a.split(",") #리스트가 됨
list_1 = []
list_2 = []
list_3 = []
```

```python
for data in a :
    if int(data)>0 and int(data)<10:
        list_1.append(data)

    elif int(data)>=10 and int(data)<100:
        list_2.append(int(data)//10)
        list_1.append(int(data) % 10)


    elif int(data) >= 100 and int(data)<1000:
        list_3.append(int(data)//100)
        list_2.append((int(data)%100)//10)
        list_1.append(int(data) % 10)
        print(list_1)
        print(list_2)
        print(list_3)
```

##문제3 다시 품

```python
a = input("1에서 1000사이에 있는 숫자를 입력해주세요")
a = a.split(",") #리스트가 됨
list_1 = []
list_2 = []
list_3 = []
```

```python
for data in a:
    if len(data) == 1 :
        list_1.append(data)
    elif len(data) == 2 :
        list_2.append(data)

    elif len(data) == 3 :
        list_3.append(data)
print(list_1)
print(list_2)
print(list_3)
```
